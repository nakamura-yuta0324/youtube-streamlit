```python
def get_data(days, tickers):
        df = pd.DataFrame()
        for company in tickers.keys():
        #company = "Meta"
            tkr = yf.Ticker(tickers[company])
            hist = tkr.history(period=f"{days}d")
            hist.index = hist.index.strftime("%d %B %Y")
            hist = hist[["Close"]]
            hist.columns = [company]
            hist = hist.T
            hist.index.name = "Name"
            hist
            df = pd.concat([df, hist])
        return df
```

```python
days = 20
tickers = {
    "apple": "AAPL",
    "Meta": "META",
    "google": "GOOGL",
    "microsoft": "MSFT",
    "netflix": "NFLX",
    "amazon": "AMZN"
}
get_data(days, tickers)
```

```python
import altair as alt
#altairはpandasとの相性が良い
```

```python
df
```

```python
companies = ["apple", "Meta"]
data = df.loc[companies]
data
```

```python
data.sort_index()
```

```python
data = data.T.reset_index()
data.head()
```

```python
data = pd.melt(data, id_vars=["Date"]).rename(
        columns={"value": "Stock Prices(USD)"}
)
data
```

```python
ymin, ymax = 150, 200

chart = (
        alt.Chart(data)
        .mark_line(opacity=0.8, clip=True) #clip=True：グラフの外にはみ出したものは消すよ
        .encode(
                x="Date:T",
                y=alt.Y("Stock Prices(USD):Q", stack=None, scale=alt.Scale(domain=[ymin, ymax])),
                #scale=alt.Scale(domain=[x,y]：x~yの範囲でグラフをスケールするよ
                color="Name:N"
        )
)
```
